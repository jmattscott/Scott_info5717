```python
from collections import Counter
from nltk.corpus import stopwords

# take second element for sort
def takeSecond(elem):
    return elem[1]

if __name__ == "__main__":
    Text = "Harry lay in his dark cupboard much later, wishing he had a watch. He didn't know what time it was and he couldn't be sure the Dursleys were asleep yet. Until they were, he couldn't risk sneaking to the kitchen for some food. He'd lived with the Dursleys almost ten years, ten miserable years, as long as he could remember, ever since he'd been a baby and his parents had died in that car crash. He couldn't remember being in the car when his parents had died. Sometimes, when he strained his memory during long hours in his cupboard, he came up with a strange vision: a blinding flash of green light and a burning pain on his forehead.  This, he supposed, was the crash, though he couldn't imagine where all the green light came from. He couldn't remember his parents at all. His aunt and uncle never spoke about them, and of course he was forbidden to ask questions. There were no photographs of them in the house. "
    word_array = Text.split(' ')
    first_vowels = []
    for word in word_array:
        if word.startswith(('a', 'e', 'i', 'o', 'u')):
            first_vowels.append(word)
    C = Counter(first_vowels)
    result = [(k, v) for k,v in C.items()]
    result.sort(key=takeSecond, reverse=True)
    print('-----Question 1. (1) Answer:-----')
    print(result)
    print('-----Question 1. (2) Answer:-----')
    for word in word_array:
        if word.find('\'') >= 0:
            print(word)
    stopwords = stopwords.words('english')
    remove_stopwords = []
    for word in word_array:
        if word in stopwords or word.find('\'') >= 0:
            continue
        remove_stopwords.append(word)
    print('-----Question 1. (3) Answer:-----')
    print(' '.join(remove_stopwords))

print("finished")
```

```python
# Question 2.1

from bs4 import BeautifulSoup
import requests
import csv
import json

def get_result(year, writer):
    for month in range(1,7):
        page = requests.post("https://www.usclimatedata.com/php/history.php", data={'location_id': 4082, 'month_nr': month, 'type': 'american', 'year_nr': year, 'weather_id': 'ustx0353'})
        response = json.loads(page.text)
        if page.status_code == 200:
            soup = BeautifulSoup(response["output"], 'html.parser')
            table = soup.find('table', {'class': 'daily_climate_table'})
            trs = table.find_all('tr')
            for idx in range(1, len(trs)):
                tds = trs[idx].find_all('td')
                date_month = tds[0].get_text().split(' ')[0] + '-' +tds[0].get_text().split(' ')[1].title()
                high = round(float(tds[1].get_text()))
                low = round(float(tds[2].get_text()))
                precip = tds[3].get_text()
                writer.writerow([date_month, high, low, precip])
                
if __name__ == "__main__":
    with open('2011-Jan-June.csv', 'w', newline='', encoding='utf-8') as file_a:
        writer1 = csv.writer(file_a, delimiter = ',')
        writer1.writerow(['Date-month', 'High', 'Low', 'Precip'])
        get_result(2011, writer1)
        file_a.close()
        
    with open('2018-Jan-June.csv', 'w', newline='', encoding='utf-8') as file_b:
        writer2 = csv.writer(file_b, delimiter = ',')
        writer2.writerow(['Date-month', 'High', 'Low', 'Precip'])
        get_result(2018, writer2)
        file_b.close()
    print("Two files were created!")
```

```python
import random

class Student:
    
    def __init__(self, first_name, middle_name, last_name, euid = '', GPA = 0, courses = []):
        self.first_name = first_name
        self.middle_name = middle_name
        self.last_name = last_name
        self.euid = euid
        self.GPA = GPA
        self.courses = courses
        
    def assign_euid(self):
        self.euid = self.first_name[0].lower() + self.middle_name[0].lower() + self.last_name[0].lower() + str(random.randint(1000, 10000))
        return '{} {} {}: {}'.format(self.first_name, self.middle_name, self.last_name, self.euid)
    
    def register(self):
        class_number = input("Enter class number: ")
        semester = input(" Enter semester: ")
        grade = float(input("Enter grade: "))
        course = Course(class_number, semester, grade)
        self.courses.append(course)
        
    def calc_gpa(self):
        sum = 0
        for course in self.courses:
            sum += course.grade
        self.GPA = round(sum/len(self.courses), 1)
        return self.GPA
    
class Course:
        
    def __init__(self, class_number, semester, grade):
        self.class_number = class_number
        self.semester = semester
        self.grade = grade
            
if __name__ == "__main__":
    
    print("-----First student-----")
    first_name1 = input("First name: ")
    middle_name1 = input("Middle name: ")
    last_name1 = input("Last name: ")
    
    print("-----Second student-----")
    first_name2 = input("First name: ")
    middle_name2 = input("Middle name: ")
    last_name2 = input("Last name: ")
  
    print("-----Third student-----")
    first_name3 = input("First name: ")
    middle_name3 = input("Middle name: ")
    last_name3 = input("Last name: ")
    
    student1 = Student(first_name1, middle_name1, last_name1)
    student2 = Student(first_name2, middle_name2, last_name2)
    student3 = Student(first_name3, middle_name3, last_name3)
    
    print(student1.assign_euid())
    print(student2.assign_euid())
    print(student3.assign_euid())
    
    student1.register()
    student2.register()
    student3.register()
    print("{} {} {}: {}".format(student1.first_name, student1.middle_name, student1.last_name, student1.calc_gpa()))
    print("{} {} {}: {}".format(student2.first_name, student2.middle_name, student2.last_name, student2.calc_gpa()))
    print("{} {} {}: {}".format(student3.first_name, student3.middle_name, student3.last_name, student3.calc_gpa()))
    
```

```python
def quiz_intro():
    print(
"""************************
** A Simple Math Quiz **
************************""")

def display_questions():
    questions = ["Addition", "Subtraction", "Multiplication", "Integer Division", "Exit"]
    for idx, element in enumerate(questions):
        print("{}. {}".format(idx+1,element))
    return questions

def section_separator():
    print("-" * 24)

def get_user_question_choice():
    i = int(input("Enter your choice: "))
    return i

def get_user_input_solution(problem):
    print("Enter your answer")
    print(problem, end="")
    result = int(input(" = "))
    return result

def solution_checking(user_solution, solution, count):
    if user_solution == solution:
        count += 1
        print('Correct.')
    else:
        print('Wrong')
    return count

def quiz_questions(index, count):
    questions = ["16 + 16", "30 - 10", "2 * 4", "6 / 3"]
    solutions = [32, 20, 8, 2]
    result = get_user_input_solution(questions[index-1])
    correct = solution_checking(result, solutions[index-1], count)
    return correct

def show_result(total, correct):
    print("You answered {} questions with {} correct".format(total, correct))
    score = round((correct / total)*100, 1)
    print("Your score is {}%. Thank you.".format(score))

def main():
    quiz_intro()
    display_questions()
    section_separator()
    option = get_user_question_choice()
    total = 0
    correct = 0
    while option != 5:
        total = total + 1
        correct = quiz_questions(option, correct)
        option = get_user_question_choice()

    print("Exit the quiz.")
    section_separator()
    show_result(total, correct)

main()
```
